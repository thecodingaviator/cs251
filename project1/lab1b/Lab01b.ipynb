```python
# Print the shape of the data
print(data.shape)

# Print the second column of the data
for i in range(0, data.shape[0]):
    print(data[i, 1])

# Print row 5 of the data
print(data[4, :])

# Print the min of each column
for i in range(0, data.shape[1]):
    print('Min of column', i, 'is', np.min(data[:, i]))

# Print the max of each row
for i in range(0, data.shape[0]):
    print('Max of row', i, 'is', np.max(data[i, :]))

# Print the mean of each column
for i in range(0, data.shape[1]):
    print('Mean of column', i, 'is', np.mean(data[:, i]))

# Print the sum of each row
for i in range(0, data.shape[0]):
    print('Sum of row', i, 'is', np.sum(data[i, :]))
```

## Task 2: MatPlotLib Basics

### 2a) Plot the data in a single plot

Using the [MatPlotLib](https://matplotlib.org/) documentation, visualize the dataset with Matplotlib. All charts should include a title, axis labels and a legend when applicable.

1. Import 'lab1.csv' into a ndarray using genfromtxt. 
2. Set figure size to 12x8. [Figure reference](https://matplotlib.org/3.1.0/api/_as_gen/matplotlib.pyplot.figure.html) [rcParams](https://matplotlib.org/3.1.0/tutorials/introductory/customizing.html#matplotlib-rcparams)
2. Create a cartesian line graph of column 2 (on the X axis) vs column 1 (Y axis) using a red dashed line. [matplotlib.pyplot introduction](https://matplotlib.org/tutorials/introductory/pyplot.html)
    1. Include a title and axis labels.

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.genfromtxt('data/lab1.csv', delimiter=',')

# Read first row of file for labels
labels = np.genfromtxt('data/lab1.csv', delimiter=',', max_rows=1, dtype=str)

# # Plot the data as 3d scatter plot
# fig = plt.figure(figsize=(12, 8))
# ax = fig.add_subplot(111, projection='3d')
# ax.scatter(data[:, 0], data[:, 1], data[:, 2])

# # Set the labels
# ax.set_xlabel(labels[0])
# ax.set_ylabel(labels[1])
# ax.set_zlabel(labels[2])

# Cartesian line graph of column 2 vs column 1 using a red line using pyplot
plt.plot(data[:, 2], data[:, 1], color='red')

# Set the labels
plt.xlabel(labels[2])
plt.ylabel(labels[1])

# Add title
plt.title('Column 2 vs Column 1')
```

### 2b) Plot the data in multiple subplots

Create 6 graphs in 2 columns and 3 rows using [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html). 
1. Row 1 is two subplots (a scatterplot and a line graph) of column 1 graphed on the x-axis VS column 0 graphed on the y-axis
2. Row 2 is two subplots (a scatterplot and a line graph) of column 2 (x-axis) VS column 1 (y-axis)
3. Row 3 is two subplots (a scatterplot and a line graph) of column 2 (x-axis) VS column 0 (y-axis)
4. Include a title and axis labels for each subplot.
5. To insure there is enough space between subplots use [subplots_adjust](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplots_adjust.html)

```python
# column 1 graphed on the x-axis VS column 0 graphed on the y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Scatterplot
ax1.scatter(data[:, 1], data[:, 0])

# Set the labels
ax1.set_xlabel(labels[1])
ax1.set_ylabel(labels[0])

# Line graph
ax2.plot(data[:, 1], data[:, 0], color='red')

# Set the labels
ax2.set_xlabel(labels[1])
ax2.set_ylabel(labels[0])

# Add title
ax1.set_title('Column 0 vs Column 1')
ax2.set_title('Column 0 vs Column 1')


# column 2 graphed on the x-axis VS column 1 graphed on the y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Scatterplot
ax1.scatter(data[:, 2], data[:, 1])

# Set the labels
ax1.set_xlabel(labels[2])
ax1.set_ylabel(labels[1])

# Line graph
ax2.plot(data[:, 2], data[:, 1], color='red')

# Set the labels
ax2.set_xlabel(labels[2])
ax2.set_ylabel(labels[1])

# Add title
ax1.set_title('Column 1 vs Column 2')
ax2.set_title('Column 1 vs Column 2')


# column 2 graphed on the x-axis VS column 0 graphed on the y-axis
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4))

# Scatterplot
ax1.scatter(data[:, 2], data[:, 0])

# Set the labels
ax1.set_xlabel(labels[2])
ax1.set_ylabel(labels[0])

# Line graph
ax2.plot(data[:, 2], data[:, 0], color='red')

# Set the labels
ax2.set_xlabel(labels[2])
ax2.set_ylabel(labels[0])

# Add title
ax1.set_title('Column 0 vs Column 2')
ax2.set_title('Column 0 vs Column 2')
```
